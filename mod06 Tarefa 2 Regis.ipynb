```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("SINASC_RO_2019.csv") 
df
```

####  1) Na base do SINASC do DataSUS, faça um boxplot da idade da mãe.

```python
fig, ax = plt.subplots()
ax.boxplot(df['IDADEMAE'])

ax.set_title('idade da mãe')
ax.set_xlabel('Idade da mãe')
ax.set_ylabel('Valor')


plt.show()
```

####  2) Faça um boxplot do peso do bebê

```python
fig, ax = plt.subplots()
ax.boxplot(df['PESO'], showfliers=True)

ax.set_title('peso do bebê')
ax.set_xlabel('Peso do bebê')
ax.set_ylabel('Valor')


plt.show()
```

####  3) Faça um histograma da idade da mãe

```python
plt.hist(df['IDADEMAE'])

plt.title('idade da mãe')
plt.xlabel('Idade da mãe')
plt.ylabel('Frequência')

plt.show()
```

#### 4) Faça um histograma do peso do bebê

```python
plt.hist(df['PESO'])

plt.title('peso do bebê')
plt.xlabel('Peso do bebê', fontsize=15)
plt.ylabel('Frequência', fontsize=15)

plt.show()
```
